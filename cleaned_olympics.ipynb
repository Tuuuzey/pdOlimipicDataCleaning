```python
import pandas as pd

bios = pd.read_csv('./bios2.csv')
```

```python
df = bios.copy()
df['Used name'] = df['Used name'].replace('•', ' ', regex=True)
df['Full name'] = df['Full name'].replace('•', ' ', regex=True)
```

```python
df = df.drop(columns=['Nick/petnames', 'Title(s)', 'Other names', 'Nationality', 'Name order', 'Original name'])
```

```python
df['height (cm)'] = df['Measurements'].str.split('/', n=1).str[0].str[:-3]
df['weight (kg)'] = df['Measurements'].str.split('/', n=1).str[1].str[:-2]
df = df.drop(columns=['Measurements'])
```

```python
date_pattern = r'(\d+ \w+ \d{4}|\d{4})'
df['Born date'] = df['Born'].str.extract(date_pattern)
df['Born date'] = pd.to_datetime(df['Born date'], format='mixed', errors='coerce')

df['Death date'] = df['Died'].str.extract(date_pattern)
df['Death date'] = pd.to_datetime(df['Death date'], format='mixed', errors='coerce')
```

```python
place_pattern = r"\d{4}\s*(?:in\s*)?(.*)"
df['Born place'] = df['Born'].str.extract(place_pattern)
df['Death place'] = df['Died'].str.extract(place_pattern)
```

```python
df.drop(columns=['Born', 'Died'])
```
